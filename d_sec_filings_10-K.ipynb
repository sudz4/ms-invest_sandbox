```python
from edgar import *
# Assuming you've already set your identity

# Get a 10-Q filing and create TenQ object
company = Company("SOFI")
filing = company.get_filings(form="10-Q").latest()
tenq = filing.obj()

# Let's see what's available
print("Available attributes and methods:")
print("-" * 40)
for item in dir(tenq):
    # Filter out private/internal attributes (those starting with _)
    if not item.startswith('_'):
        # Get the type of the attribute
        attr_type = type(getattr(tenq, item))
        print(f"{item}: {attr_type.__name__}")

# If you want to see available financial statement methods specifically:
print("\nFinancial statement methods:")
print("-" * 40)
fin = tenq.financials
for item in dir(fin):
    if not item.startswith('_'):
        print(item)

# And to see what items (sections) are available in this specific filing:
print("\nAvailable items/sections in this filing:")
print("-" * 40)
for item in tenq.items:
    print(item)
```
